```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sample_df = pd.read_csv("C:/MyWorkspace/Python_MachineLearning/Clustering/flame.csv")
```

```python
training_points = sample_df[["col1", "col2"]]
training_labels = sample_df["target"]
```

```python
kmeans = KMeans(n_clusters=2).fit(training_points)
```

```python
plt.scatter(training_points["col1"], training_points["col2"], c=kmeans.labels_, cmap='rainbow')
plt.show()
```

```python
agglo = AgglomerativeClustering(n_clusters = 2).fit(training_points)
```

```python
plt.scatter(training_points["col1"], training_points["col2"], c = agglo.labels_, cmap = 'rainbow')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.93, min_samples=4).fit(training_points)
```

```python
plt.scatter(training_points["col1"], training_points["col2"], c=dbscan.labels_, cmap='rainbow')
plt.show()
```
